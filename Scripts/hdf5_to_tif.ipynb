```python
topdirectory = os.listdir("PSI/Novembre_2021")

for element in topdirectory[1:]:
    middirectory = (os.listdir("PSI/Novembre_2021/"+element))
    for folder in middirectory:
        if folder != ".DS_Store":
            titlearray=os.listdir("PSI/Novembre_2021/"+element+"/"+folder)
            
            for title in titlearray:
                information = h5py.File("PSI/Novembre_2021/"+element+"/"+folder+"/"+title,"r")["entry"].get("data").get("data")
                imageinfo = np.zeros([1024,1024])
                for i in range(1024):
                    imageinfo[i] = information[i]
                    Image.fromarray(imageinfo).save("PSI/Novembre_2021/"+element+"_tif/"+folder+"/"+title[:-4]+".tif")
            
            print(element,folder)
```

```python
middirectory = os.listdir("PSI/Novembre_2021/Position_1")
for folder in middirectory:
    if folder != ".DS_Store":
        titlearray=os.listdir("PSI/Novembre_2021/Position_1"+"/"+folder)

        for title in titlearray:
            information = h5py.File("PSI/Novembre_2021/Position_1"+"/"+folder+"/"+title,"r")["entry"].get("data").get("data")
            imageinfo = np.zeros([1024,1024])
            for i in range(1024):
                imageinfo[i] = information[i]
                Image.fromarray(imageinfo).save("PSI/Novembre_2021/Position_1"+"_tif/"+folder+"/"+title[:-4]+".tif")
                
        print("Position 1",folder)
```
