```python
histogram = np.zeros(40000)
Foldername = "ResultsPavia/"
folderrange = len(os.listdir(Foldername))
#for fileindex in range(int(folderrange)):
preanalisis = []
for fileindex in range(folderrange):
    title = Foldername + str(fileindex)+".txt"
    with open(title, 'r') as file:
        position = np.array(eval(file.read())) # read list string and convert to array
        histogram += position[:40000]
        file.close()
hist_reshaped = np.reshape(histogram,[200,200])
```

```python
fig = plt.figure(figsize=(14,10))
mat = fig.add_subplot(111)
tax = mat.matshow(hist_reshaped)
plt.title("Intensity graph with LD=160")
plt.xlabel("X coordinate (a.u.)")
plt.ylabel("Y coordinate (a.u.)")
#plt.imshow(tax)
fig.colorbar(tax,label="Counts")
```

```python
i=0
histogram_theta = []
histogram_phi = []
ResultsFolder = "ResultsPavia/"
amount = len(content)
#for elem in os.listdir("ResultsPavia/")[len(preanalisis):]:
for elem in content:
    print("iteration: ",i," out of ",len(content))
    if i< amount/2:
        with open(ResultsFolder+elem, 'r') as file:
            position = np.array(eval(file.read())) # read list string and convert to array
            for q in position:
                histogram_phi.append(q)
            file.close()       
    else:
        with open(ResultsFolder+elem, 'r') as file:
            position = np.array(eval(file.read())) # read list string and convert to array
            for q in position:
                histogram_theta.append(q)
            file.close()      
    i+=1
fig , (ax1,ax2) = plt.subplots(1,2)
fig.set_figheight(5)
fig.set_figwidth(15)
ax1.hist(histogram_phi,200)
ax1.set_title("Phi distribution (rad)")
ax1.set_xlabel("Rad")
ax1.set_ylabel("Counts")
ax2.hist(histogram_theta,200,range=[0,0.04])
ax2.set_title("Theta distribution (rad)")
ax2.set_xlabel("Rad")
ax2.set_ylabel("Counts")
#plt.hist(histogram_phi)
#plt.hist(histogram_theta)
```

```python
def format_radians_label(float_in):
    # Converts a float value in radians into a
    # string representation of that float
    string_out = str(float_in / (np.pi))+"π"
    
    return string_out
def convert_polar_xticks_to_radians(ax):
    # Converts x-tick labels from degrees to radians
    
    # Get the x-tick positions (returns in radians)
    label_positions = ax.get_xticks()
    
    # Convert to a list since we want to change the type of the elements
    labels = list(label_positions)
    
    # Format each label (edit this function however you'd like)
    labels = [format_radians_label(label) for label in labels]
    
    ax.set_xticklabels(labels)
print(np.max(histogram_theta))
print(np.average(histogram_theta))
print(np.average(histogram_phi))
print(1/np.tan(np.max(histogram_theta)/2))

#bins = binss*180/math.pi
fig = plt.figure()
fig.set_figheight(7)
fig.set_figwidth(25)
ax1 = plt.subplot(1,3,2)
ax1.hist(histogram_phi,200)
ax1.set_title("Phi distribution (rad)")
ax1.set_xlabel("Rad")
ax1.set_ylabel("Counts")
ax2 = plt.subplot(1,3,1)
ax2.hist(histogram_theta,bins=200,range=[0,0.038])
ax2.set_title("Theta distribution (rad)")
ax2.set_xlabel("Rad")
ax2.set_ylabel("Counts")
ax3 = plt.subplot(133,projection="polar")
ax3.bar(bins[:200],values,width=0.05,bottom=0.0,align="center")
ax3.set_ylim([0,np.max(values)+1])
convert_polar_xticks_to_radians(ax3)
```
