```python
import numpy as np
import matplotlib.pyplot as plt 
import random
import math
import time
from mpl_toolkits.mplot3d import Axes3D
import os
```

```python
def plot3D(x,y,dist_to_det,condition_white,condition_ax):
    hist,xax,yax,image = plt.hist2d(x,y,bins=200,range=[[10,30],[10,30]])
    print("file(s) saved")
    return hist
```

```python
Foldername = "Test6/"
IndexX = 90
IndexY = 64
finalhist = np.zeros([200,200])
bins = np.zeros(20)
Foldername = "Pavia7/"
folderrange = len(os.listdir(Foldername))
for fileindex in range(5):
    print("Fileindex number "+str(fileindex))
    title = Foldername+"Testrun-of-full-sim-positionWW-index"+str(fileindex)+".txt"
    i,run = 0,0
    with open(title, 'r') as file:
        position = np.array(eval(file.read())) # read list string and convert to array
        file.close()
    hist = plot3D(position[:,0],position[:,1],fileindex,True,True)
    finalhist += hist
    print("Finalhist upgraded, round ",fileindex)
```

```python
fig = plt.figure(figsize=(14,10))
mat = fig.add_subplot(111)
tax = mat.matshow(finalhist)
#plt.imshow(tax)
fig.colorbar(tax)
```
